```python
# Import Pandas library and Read 'Appt_No_Show.csv' file

import pandas as pd

no_show = pd.read_csv('Appt_No_Show.csv')
```

```python
# Quick review on dataset
no_show.info()
```

```python
# Create a copy file before cleaning up the original dataset. This will help generate a cleaned version while maintaining the original file
no_show_cleaned = no_show.copy()
```

```python
# Check to see if there is duplicate
dup = no_show_cleaned.duplicated().sum()
print(dup)

# Confirmed that no duplicate exists
```

```python
# Find minimum age and maximum age
min_age = no_show_cleaned['Age'].min()
max_age = no_show_cleaned['Age'].max()

print("Minimum Age:", min_age)
print("Maximum Age:", max_age)

# Found some outliers (Age <0 and Age >100)
```

```python
# Exclude the age outliers from dataset
no_show_cleaned = no_show_cleaned[(no_show_cleaned['Age']>0) & (no_show_cleaned['Age']<100)]
no_show_cleaned.info()
```

```python
# Convert ScheduledDay and AppointmentDay columns to date type
date_cols = ["ScheduledDay", "AppointmentDay"]
for col in date_cols:
    no_show_cleaned[col] = pd.to_datetime(no_show_cleaned[col], errors="coerce").dt.normalize()
```

```python
# Confirm the data type change for ScheduledDay and AppointmentDay
no_show_cleaned.head()
```

```python
# Creat a new columm for WaitDays
no_show_cleaned["WaitDays"] = (no_show_cleaned["AppointmentDay"] - no_show_cleaned["ScheduledDay"]).dt.days
no_show_cleaned.head(10)
```

```python
# Find any outlier in WaitDays. WaitDays should be always equal to 0 or greater than 0
no_show_cleaned['WaitDays'].min()
```

```python
# Exclude the WaitDays outliers from dataset
no_show_cleaned = no_show_cleaned[(no_show_cleaned['WaitDays']>=0)]
no_show_cleaned.info()
```

```python
# Modify column names for no_show, hypertension, and handicap

no_show_cleaned = no_show_cleaned.rename(columns={"No-show":"No_show"})
no_show_cleaned = no_show_cleaned.rename(columns={"Hipertension":"Hypertension"})
no_show_cleaned = no_show_cleaned.rename(columns={"Handcap":"Handicap"})

no_show_cleaned.head()
```

```python
# Convert No_show from Yes/No to 1/0 for analysis
no_show_cleaned['No_show'] = no_show_cleaned['No_show'].replace({'Yes':1, 'No':0})
no_show_cleaned['No_show'] = no_show_cleaned['No_show'].astype(int)
no_show_cleaned.head(5)
```

```python
no_show_cleaned.info()
```

```python
# Find any correlation between different factors and no_show pattern using Pearson correlation coefficient

corr_scholarship = no_show_cleaned["Scholarship"].corr(no_show_cleaned["No_show"])
corr_SMS = no_show_cleaned["SMS_received"].corr(no_show_cleaned["No_show"])
corr_HTN = no_show_cleaned["Hypertension"].corr(no_show_cleaned["No_show"])
corr_DM = no_show_cleaned["Diabetes"].corr(no_show_cleaned["No_show"])
corr_Alcohol = no_show_cleaned["Alcoholism"].corr(no_show_cleaned["No_show"])
corr_Handicap = no_show_cleaned["Handicap"].corr(no_show_cleaned["No_show"])
print("Correlation between scholarship and No_Show:", corr_scholarship)
print("Correlation between SMS_received and No_Show:", corr_SMS)
print("Correlation between Hypertension and No_Show:", corr_HTN)
print("Correlation between Diabetes and No_Show:", corr_DM)
print("Correlation between Alcoholism and No_Show:", corr_Alcohol)
print("Correlation between Handicap and No_Show:", corr_Handicap)
```

```python
# Find any correlation between Gender and no_show pattern using Chi-square test

import pandas as pd
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(
    no_show_cleaned['Gender'],
    no_show_cleaned['No_show']
)

# Run Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)
```
